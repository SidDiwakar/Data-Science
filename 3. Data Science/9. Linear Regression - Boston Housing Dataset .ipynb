# Linear Regression : Boston Housing Dataset 



```python

import sklearn
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(10)
from sklearn.datasets import load_boston
import seaborn as sns; sns.set(color_codes=True)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Loading the dataset and simple analysis

boston = load_boston()

print(boston.data.shape)
print(boston.data)
print(boston.feature_names)
print(boston.target)
```

```python
boston_dataframe = pd.DataFrame(boston.data)
boston_dataframe.head()

boston_dataframe['Price'] = boston.target

x = boston_dataframe.drop('Price', axis = 1)
y = boston_dataframe['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= .3 , random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lm = LinearRegression()
lm.fit(x_train, y_train)

y_pred = lm.predict(x_test)
plt.scatter(y_test, y_pred)
plt.xlabel("Prices : $Y_i$")
plt.ylabel("Predicted_prices : $\hat{Y}_i$")
plt.title("Price vs Predicted prices : $Y_i$ vs $\hat{Y}_i$")
plt.show()
```

```python
delta_y = y_test - y_pred;

sns.set_style('whitegrid')
sns.kdeplot(delta_y)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.kdeplot(np.array(y_pred), bw=0.5)
plt.show()
```
