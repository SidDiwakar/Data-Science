# Generating Mock Data with Python! (NumPy, Pandas, & Datetime Libraries)

```python
import pandas as pd
import datetime 
import numpy
import random
```

```python
Products = ["Address books","Agendas","All holidays", "Animation and 3D","Announcements","Autumn", "Baby","Back to school","Budgets"
     "Bids and quotes" , "Birthday","Blank and general","Brochures","Business cards","Business plans", "Calendars","Cards","Certificates",
     "Charts","Christmas"]
   
Price = []
for x in range(0,len(Products)):
        n = random.randint(500,600)
        Price.append(n)
        
# print(Price)

Product_dict = dict(zip(Products,Price))
print(Product_dict)
```

```python
columns = ['Order ID', 'Products', 'Quantity ordered', 'Price each', 'Ordered date', 'Purchase Address']

df = pd.DataFrame(columns=columns)

for i in range(1000):
    Product = random.choice(list(Product_dict.keys()))
    price = Product_dict[Product]
    df.loc[i] = [i, Product, random.randint(1,5), price, 'NA', 'NA']

df.to_csv('test_data.csv')
```
